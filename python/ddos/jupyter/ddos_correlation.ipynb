```python
import numpy as np
import pandas as pd


def get_orgrin_data(file_path):
    return pd.read_csv(file_path, low_memory=False)


def get_data_set(file_path):

    ddos_data = pd.read_csv(file_path, low_memory=False)
    # 暂时剔除无意义的特征
    del ddos_data['Row']
    # 删除非数值列
    other_data = ddos_data.drop([
        'Flow ID', 'Src IP', 'Src Port', 'Dst IP', 'Dst Port', 'Protocol',
        'Timestamp'
    ],
                                axis=1)
    # 删除空值
    other_data = other_data.dropna(how='all')
    # 转换为numpy  aarray
    new_data = other_data.to_numpy()
    # 拆分特征列和目标列
    X, y = new_data[:, :-1], new_data[:, -1]
    # 数值型转换类型
    X = X.astype(float)
    # 处理nan or inf
    X = np.nan_to_num(X)
    return X, y
```

```python
file_path="/work/datasets/unbalaced_test.csv"
ddos_data = get_orgrin_data(file_path)
ddos_data.head()
```

```python
ddos_data["Label"]=ddos_data["Label"].map({"Benign":0,"ddos":1})
ddos_data
```

```python
X_14=ddos_14.to_numpy().astype(float)
```

```python
def label_encoder(y):
    # 标签数值化，方便处理
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    return le.fit_transform(y)
```

```python
_, y = get_data_set(file_path)
y=label_encoder(y)
y
```

```python
y=np.hstack((y,1))
```

```python
X_14.shape
```

```python
#### 决策树验证一下准确率
from sklearn.model_selection import train_test_split
X_14_train, X_14_test, y_train, y_test = train_test_split(X_14,
                                                          y,
                                                          test_size=0.2,
                                                          random_state=2019)
```

```python
%%time
from sklearn.tree import DecisionTreeClassifier
dt_tree = DecisionTreeClassifier()
dt_tree.fit(X_14_train, y_train)
y_pred = dt_tree.predict(X_14_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
